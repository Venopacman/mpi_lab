```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = [15,8]
```

```python
fp_local_df = pd.read_csv('fp_experiment_cluster.csv')
```

```python
fp_local_df.Algorithm.unique()
```

```python
gauss_df = fp_local_df[fp_local_df.Algorithm == 'gauss.py']
rectangular_df = fp_local_df[fp_local_df.Algorithm == 'rectangular.py']
simpson_df = fp_local_df[fp_local_df.Algorithm == 'simpson.py']
trapesoidal_df = fp_local_df[fp_local_df.Algorithm == 'trapesoidal.py']
```

```python
plt.plot(gauss_df.Cores, gauss_df.AVG_time , 'go')
plt.plot(rectangular_df.Cores, rectangular_df.AVG_time, 'ro')
plt.plot(simpson_df.Cores, simpson_df.AVG_time, 'bo')
plt.plot(trapesoidal_df.Cores, trapesoidal_df.AVG_time, 'yo')
plt.ylabel('Average computing time')
plt.xlabel('# of threads')
plt.title('Average time vs amount of threads')
plt.legend(fp_local_df.Algorithm.unique(),)
plt.show()
```
