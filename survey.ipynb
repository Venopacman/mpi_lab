```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = [15,8]
```

```python
fp_cluster_df = pd.read_csv('fp_experiment_cluster.csv')
```

```python
fp_cluster_df.Algorithm.unique()
```

```python
gauss_df = fp_cluster_df[fp_cluster_df.Algorithm == 'gauss.py']
rectangular_df = fp_cluster_df[fp_cluster_df.Algorithm == 'rectangular.py']
simpson_df = fp_cluster_df[fp_cluster_df.Algorithm == 'simpson.py']
trapesoidal_df = fp_cluster_df[fp_cluster_df.Algorithm == 'trapesoidal.py']
```

```python
plt.plot(gauss_df.Cores, gauss_df.AVG_time , 'g--')
plt.plot(rectangular_df.Cores, rectangular_df.AVG_time, 'r-')
plt.plot(simpson_df.Cores, simpson_df.AVG_time, 'b-.')
plt.plot(trapesoidal_df.Cores, trapesoidal_df.AVG_time, 'y:')
plt.ylabel('Average computing time')
plt.xlabel('# of threads')
plt.title('Average time vs amount of threads on server')
plt.legend(fp_cluster_df.Algorithm.unique(),)
plt.show()
```
